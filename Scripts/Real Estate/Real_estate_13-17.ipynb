```python
def plot_pie_charts(df):
    years = df['Year']
    sectors = df.columns[1:]
    
    # Number of years
    n_years = len(years)
    
    # Number of rows needed
    n_rows = n_years // 3 + (1 if n_years % 3 != 0 else 0)
    
    # Create subplots with multiple rows
    fig, axes = plt.subplots(n_rows, 3, figsize=(15, 5 * n_rows))
    
    for i, year in enumerate(years):
        row_idx = i // 3
        col_idx = i % 3
        
        ax = axes[row_idx, col_idx] if n_rows > 1 else axes[col_idx]
        sizes = df.loc[i, sectors]
        
        # Check for non-negative values
        if any(sizes < 0):
            print(f"Skipping year {year} due to negative values")
            continue
        
        ax.pie(sizes, labels=sectors, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
        ax.set_title(f'Year {year}')
    
    plt.tight_layout()
    plt.show()

# Example usage
plot_pie_charts(df)
```

```python
def convert_to_percentage(df):
    # Select only the numeric columns (excluding the 'Year' column)
    numeric_columns = df.columns.difference(['Year'])
    
    # Convert values to percentages
    df[numeric_columns] = df[numeric_columns].div(df[numeric_columns].sum(axis=1), axis=0) * 100
    
    return df

# Convert values to percentages
df_percentage = convert_to_percentage(df)
print(df_percentage)
```

```python
def plot_horizontal_stacked_bar_chart(df, filename):
    # Convert values to percentages
    df = convert_to_percentage(df)
    
    # Set the figure size
    fig, ax = plt.subplots(figsize=(10, 6))

    colors = ['red', 'blue', 'green', 'orange']
    
    # Plot the horizontal stacked bar chart
    bars = df.set_index('Year').plot(kind='barh', stacked=True, color=colors, width=0.7, ax=ax)

    # Add labels and title
    ax.set_xlabel('Percentage')
    ax.set_ylabel('Year')
    ax.set_title('Percentage of Each Sector Over Years')

    # Add percentages on the bars
    for bar in bars.patches:
        width, height = bar.get_width(), bar.get_height()
        x, y = bar.get_xy() 
        ax.text(x + width / 2, y + height / 2, f'{width:.1f}%', ha='center', va='center',color='white',weight='bold')

    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    
    plt.savefig(filename)
    
    # Show the plot
    plt.show()

# Example usage
plot_horizontal_stacked_bar_chart(df_percentage, "..\..\Saved figures\Real estate\Funds.png")
```

```python
def process_numbers(paragraph):
    # Split the paragraph into individual numbers
    numbers = paragraph.split()
    
    # Initialize a list to store the result lines
    result_lines = []
    
    # Process the numbers in chunks of 7
    for i in range(0, len(numbers), 6):
        # Extract the current chunk of 7 numbers
        chunk = numbers[i:i + 6]
        
        # If there are at least two numbers in the chunk, take the first two
        if len(chunk) >= 2:
            result_lines.append(f"{chunk[0]} {chunk[1]}")
    
    # Join the result lines with newline characters and return
    return "\n".join(result_lines)

# Example usage
paragraph = updated_paragraph
result = process_numbers(paragraph)
# print(result)
with open("..\..\Data\Preprocessed\preprocessed_txt\Real_estate\E19_12.txt", "w") as file:
    file.write(result)

print("Result has been saved to E19_12.txt")
```

```python
def create_dataframe_from_txt(file_path):
    # Initialize empty lists to store the data
    years = []
    price_list = []

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            # Clean and split the line into parts using regex to handle different separators
            parts = re.split(r'\s+|[-]', line.strip())
            if len(parts) == 2:
                try:
                    # Convert the parts to integers and append to the lists
                    year = int(parts[0])
                    price = int(parts[1])
                    years.append(year)
                    price_list.append(price)
                except ValueError:
                    # Handle the case where conversion to int fails
                    print(f"Skipping line due to conversion error: {line.strip()}")
    
    # Create the DataFrame
    data = {'Year': years, 'Average Selling price of Commercial Buildings': price_list}
    df = pd.DataFrame(data)
    
    return df

# Example usage
file_path = '..\..\Data\Preprocessed\preprocessed_txt\Real_estate\E19_12.txt'
df = create_dataframe_from_txt(file_path)
print(df)
```

```python
def plot_line_chart(df, filename):
    # Plot the line chart
    plt.figure(figsize=(15, 6))
    plt.plot(df['Year'], df['Average Selling price of Commercial Buildings'], marker='o', linestyle='-', color='b')
    
    # Set the ticks on the x-axis to be the specific years in the DataFrame
    plt.xticks(df['Year'])
    
    plt.xlabel('Year')
    plt.ylabel('Price (yuan/sq.m)')
    plt.title('Average Selling price of Commercial Buildings')
    plt.grid(True)
    
    # Save the plot as a PNG file
    plt.savefig(filename)
    
    plt.show()

# Example usage
# file_path = 'path_to_your_file.txt'
# df = create_dataframe_from_txt(file_path)
# print(df)
plot_line_chart(df, "..\..\Saved figures\Real estate\House_price.png")
```
