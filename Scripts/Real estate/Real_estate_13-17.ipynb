### Image E19-13

```python
from decimal import Decimal

def create_dataframe_from_txt(file_path):
    """
    Create a DataFrame from a text file containing year and number of enterprises data.
    Input:
        file_path: str - path to the text file 
    Output:
        df: pd.DataFrame - DataFrame containing the year and number of enterprises data 
    """
    # Initialize empty lists to store the data
    years = []
    num_enterprises_list = []

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            # Clean and split the line into parts using regex to handle different separators
            parts = re.split(r'\s+|[-]', line.strip())
            if len(parts) == 2:
                try:
                    # Convert the parts to integers and append to the lists
                    year = int(parts[0])
                    num_enterprises = Decimal(parts[1])  # Convert string to decimal
                    years.append(year)
                    num_enterprises_list.append(num_enterprises)
                except ValueError:
                    # Handle the case where conversion to int or decimal fails
                    print(f"Skipping line due to conversion error: {line.strip()}")
    
    # Create the DataFrame
    data = {'Year': years, 'Ratio of Liabilities to Assets (%)': num_enterprises_list}
    df = pd.DataFrame(data)
    
    return df

# Example usage
file_path = '..\..\Data\Preprocessed\preprocessed_txt\Real_estate\E19_13.txt'
df = create_dataframe_from_txt(file_path)
print(df)
```

#### Plot line chart

```python
def plot_line_chart(df, filename):
    """
    Plot a line chart from the DataFrame and save it as a PNG file.
    Input:
        df: pd.DataFrame - DataFrame containing the data to plot
        filename: str - name of the output PNG file
    Output:
        line chart saved as a PNG file
    """
    # Plot the line chart
    plt.figure(figsize=(15, 6))
    plt.plot(df['Year'], df['Ratio of Liabilities to Assets (%)'], marker='o', linestyle='-', color='b')
    
    # Set the ticks on the x-axis to be the specific years in the DataFrame
    plt.xticks(df['Year'])
    
    plt.xlabel('Nm')
    plt.ylabel('T l n phi tr trn ti sn (%)')
    plt.title('Ti sn v n ca cc cng ty bt ng sn Trung Quc')
    plt.grid(True)
    
    # Save the plot as a PNG file
    plt.savefig(filename)
    
    plt.show()

plot_line_chart(df, "..\..\Saved figures\Real estate\Ratio_of_Liabilities.png")
```

### Image E19-16

```python
def create_dataframe_from_txt(file_path):
    """
    Create a DataFrame from a text file containing year and number of completed and sold flats.
    Input:
        file_path: str - path to the text file
    Output:
        df: pd.DataFrame - DataFrame containing the year and number of completed and sold flats
    """
    # Initialize empty lists to store the data
    years = []
    pending_list = []
    purchased_list = []

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            # Clean and split the line into parts using regex to handle different separators
            parts = re.split(r'\s+|[-]', line.strip())
            if len(parts) == 3:
                try:
                    # Convert the parts to integers and append to the lists
                    year = int(parts[0])
                    pending = int(parts[1])
                    purchased = int(parts[2])
                    years.append(year)
                    pending_list.append(pending)
                    purchased_list.append(purchased)
                except ValueError:
                    # Handle the case where conversion to int fails
                    print(f"Skipping line due to conversion error: {line.strip()}")
    
    # Create the DataFrame
    data = {'Year': years, 'Total Number of Flats of Residential Buildings Completed': pending_list, 'Total Number of Flats of Residential Buildings Sold': purchased_list}
    df = pd.DataFrame(data)
    
    return df

# Example usage
file_path = '..\..\Data\Preprocessed\preprocessed_txt\Real_estate\E19_16.txt'
df = create_dataframe_from_txt(file_path)
print(df)
```

```python
def plot_grouped_bar_chart(df, filename):
    """
    Plot a grouped bar chart from the DataFrame and save it as a PNG file.
    Input:
        df: pd.DataFrame - DataFrame containing the data to plot
        filename: str - name of the output PNG file
    Output:
        grouped bar chart saved as a PNG file
    """
    # Set the positions and width for the bars
    bar_width = 0.35
    bar_positions = np.arange(len(df['Year']))

    # Create the plot
    plt.figure(figsize=(20, 10))

    # Plot the bars for 'Area of Land Pending for Development'
    plt.bar(bar_positions - bar_width/2, df['Total Number of Flats of Residential Buildings Completed'], width=bar_width, color='blue', label='Total Number of Flats of Residential Buildings Completed')

    # Plot the bars for 'Area of Land Purchased in the year'
    plt.bar(bar_positions + bar_width/2, df['Total Number of Flats of Residential Buildings Sold'], width=bar_width, color='orange', label='Total Number of Flats of Residential Buildings Sold')

    # Add labels, title, and legend
    plt.xlabel('Year')
    plt.ylabel('Flat')
    plt.title('Number of Flat of Residential Buildings Completed and Sold Over by Enterprises in Real Estate Development')
    plt.xticks(bar_positions, df['Year'])
    plt.legend()

    # Show the plot
    plt.tight_layout()
    
    # # Save the plot as a PNG file
    plt.savefig(filename)
    
    plt.show()

# Example usage
plot_grouped_bar_chart(df, "..\..\Saved figures\Real estate\Flat.png")
# plot_line_chart(df, "..\..\Saved figures\Real estate\Employed_foreign.png")
```
