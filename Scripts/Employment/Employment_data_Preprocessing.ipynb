```python
import numpy as np
import pandas as pd
import os
from typing import List
import easyocr
import re
import requests
import shutil
from spellchecker import SpellChecker
import matplotlib.pyplot as plt
```


## Employment - Table 4.7 - Column: Total_wage_bill_100_million_yuan

### Read from txt file to create pandas data frame

```python
def create_dataframe_from_txt(file_path):
    # Initialize empty lists to store the data
    years = []
    total_wage_bill = []

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            # Clean and split the line into parts using regex to handle different separators
            parts = re.split(r'\s+|[-]', line.strip())
            if len(parts) == 2:
                try:
                    # Convert the parts to integers and append to the lists
                    year = int(parts[0])
                    num_enterprises = int(parts[1])
                    years.append(year)
                    total_wage_bill.append(num_enterprises)
                except ValueError:
                    # Handle the case where conversion to int fails
                    print(f"Skipping line due to conversion error: {line.strip()}")

    # Create the DataFrame
    data = {'Year': years, 'Total wage bill 100 million yuan': total_wage_bill}
    df = pd.DataFrame(data)

    return df

# Example usage
file_path = '/content/Total_wage_bill_100_million_yuan.txt'
df = create_dataframe_from_txt(file_path)
print(df)
```

### Plot line chart

```python
def plot_line_chart(df):
    # Plot the line chart
    plt.figure(figsize=(15, 6))
    plt.plot(df['Year'], df['Total wage bill 100 million yuan'], marker='o', linestyle='-', color='b')

    # Set the ticks on the x-axis to be the specific years in the DataFrame
    plt.xticks(df['Year'])

    plt.xlabel('Year')
    plt.ylabel('Total wage bill')
    plt.title('Total wage bill Over Years')
    plt.grid(True)

    # Save the plot as a PNG file
    # plt.savefig(filename)

    plt.show()

# Example usage
# file_path = 'path_to_your_file.txt'
# df = create_dataframe_from_txt(file_path)
# print(df)

plot_line_chart(df)
```

## Employment - China unemployment rate. [Link](https://tradingeconomics.com/china/unemployment-rate)

### Read from txt file to create pandas data frame

```python
def create_dataframe_from_txt(file_path):
    # Initialize empty lists to store the data
    years = []
    unemployment_rate = []

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            # Clean and split the line into parts using regex to handle different separators
            parts = re.split(r'\s+|[-]', line.strip())
            if len(parts) == 2:
                try:
                    # Convert the parts to integers and append to the lists
                    year = int(parts[0])
                    rate = float(parts[1])
                    years.append(year)
                    unemployment_rate.append(rate)
                except ValueError:
                    # Handle the case where conversion to int fails
                    print(f"Skipping line due to conversion error: {line.strip()}")

    # Create the DataFrame
    data = {'Year': years, 'Unemployment rate': unemployment_rate}
    df = pd.DataFrame(data)

    return df

# Example usage
file_path = '/content/Surveyed_unemployment_rate.txt'
df = create_dataframe_from_txt(file_path)
print(df)
```
